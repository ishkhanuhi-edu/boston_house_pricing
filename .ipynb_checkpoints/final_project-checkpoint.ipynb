## Abstract:
The problem that we are going to solve here is that given a set of features that describe a house in Boston, our machine learning model must predict the house price.

## About the dataset

The dataset is collected from [Kaggle](https://www.kaggle.com/competitions/boston-housing/data). Let's get into the data and know more about it.

This dataset contains information collected by the U.S Census Service concerning housing in the area of Boston Mass. It was obtained from the StatLib archive Data, and has been used extensively throughout the literature to benchmark algorithms. However, these comparisons were primarily done outside of Delve and are thus somewhat suspect. The dataset is small in size with only 506 cases.

The data was originally published by Harrison, D. and Rubinfeld, D.L. Hedonic prices and the demand for clean air', J. Environ. Economics & Management, vol.5, 81-102, 1978.

Variables There are 14 attributes in each case of the dataset. They are:

ZN - proportion of residential land zoned for lots over 25,000 sq.ft.

INDUS - proportion of non-retail business acres per town.

CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)

NOX - nitric oxides concentration (parts per 10 million)

RM - average number of rooms per dwelling

AGE - proportion of owner-occupied units built prior to 1940

DIS - weighted distances to five Boston employment centres

RAD - index of accessibility to radial highways

TAX - full-value property-tax rate per 10,000 dollars.

PTRATIO - pupil-teacher ratio by town

B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town

LSTAT - % lower status of the population

MEDV - Median value of owner-occupied homes in 1000's dollars

## Importance of Predicting Boston House Prices

Predicting house prices, such as those in Boston, is a crucial aspect of real estate economics and serves several important functions:

1. **Market Understanding**: Accurately predicting house prices can help buyers, sellers, and real estate professionals understand the current state of the market, including its trends, and make more informed decisions.
2. **Investment Decisions**: For investors, an accurate prediction of house prices is key to making profitable investments. Investors can decide when to buy or sell properties based on the predicted prices and trends.
3. **Policy Making**: Local and state governments, as well as housing authorities, can use housing price predictions to inform housing and urban development policies, affordable housing strategies, and property tax calculations.
4. **Lending Decisions**: Banks and mortgage companies can use predicted prices to assess the value of the house, which could be used as collateral for a mortgage loan. This can influence the amount that can be borrowed and the interest rate.
5. **Insurance Industry**: Insurance companies can use house price predictions to determine the level of homeowners insurance that is needed.

The Boston Housing dataset, in particular, has been widely used in machine learning and statistics to demonstrate and teach predictive modeling techniques. Its popularity also comes from its relatively small size and the clear problem definition, which makes it a good dataset for beginners to learn on. However, it is worth noting that the dataset is quite old (from 1978), and the housing market may have changed significantly since then.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python

# Initializing column names
columns = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptration', 'b', 'lstat', 'medv']

# Loading Boston Housing Dataset
data = pd.read_csv('data/boston.csv', delimiter=r"\s+", names = columns, index_col=False)
```

```python
from scipy import stats


for column in data.columns:
    plt.figure(figsize=(12,6))

    plt.subplot(1, 2, 1)
    sns.histplot(data[column], bins=30)
    plt.title('Histogram')

    plt.subplot(1, 2, 2)
    stats.probplot(data[column], plot=plt)
    plt.title('Q-Q Plot')

    plt.suptitle('Checking Normality of '+ column, size=14)
    plt.show()
```

### Conclusion

From the above plots, we can conclude that only the variable 'rm' seems to follow a normal distribution, as its Q-Q plot falls approximately on a straight line. Deviations from this straight line suggest departures from normality.

## Building Linear Regression Model

```python
# Arranging features based on features importance
features_arranged_on_importance = important_features.sort_values(ascending = False).index
features_arranged_on_importance
```

```python
y = data.loc[:, 'medv'].values
```

```python
# Arranging columns based on features importance
new_boston = data[features_arranged_on_importance]
new_boston.head()
```

```python
# Getting boston values
X = new_boston.values
X = X[:, :6]

# TODO : Splitting data as train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

print('Training Score : ', linear_model.score(X_train, y_train))
print('Testing Score  : ', linear_model.score(X_test, y_test))

print('R2 Score : ', r2_score(y_test, linear_model.predict(X_test)))
print('MSE : ', mean_squared_error(y_test, linear_model.predict(X_test)))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, linear_model.predict(X_test))))

```

**Training Score**: This is the coefficient of determination R² of the prediction for our training set. It represents the proportion of the variance for the dependent variable that's explained by the independent variables. In our case, it's about 0.69 or 69%, which means our model explains 69% of the variability in the house prices in the training dataset.

**Testing Score**: This is the R² score for our test dataset, which is about 0.62 or 62%. This means that our model explains 62% of the variability in the house prices in the test dataset. It's lower than the training score, which is quite common as models tend to perform slightly worse on new data.

**R² Score**: This is the same as our testing score since it is calculated on our test set. Again, it's 62%.

**MSE (Mean Squared Error)**: This is a measure of how well the model generalizes to new data. The smaller the Mean Squared Error, the closer the fit is to the data. The MSE of 27.8 is the average squared difference between the estimated values and what is estimated. High MSE implies that our model still makes sizable errors in prediction.

**RMSE**: This value gives the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are, and RMSE is a measure of how spread out these residuals are. In other words, it tells you how concentrated the data is around the line of best fit.

So, our RMSE value is, 5.27, it means that our model's predictions are, on average, approximately 5270 US dollars away from the actual value. Given that the prices in the Boston Housing dataset range from 5000 to 50000 US dollars, we can make a judgement on whether this error is acceptable or not based on our specific requirements or thresholds.



**In conclusion**, our model is doing an okay job predicting house prices - not perfect, but it has demonstrated the capability to capture some patterns in the data. It seems to perform better on the training data compared to the test data, suggesting some degree of overfitting. The MSE indicates there's still room for improvement. To increase the model performance, we can consider feature engineering, using a more complex model, or including regularization techniques to avoid overfitting in the future.


## PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)


pca = PCA()
principalComponents1 = pca.fit_transform(data_scaled)

plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Boston Housing Dataset Explained Variance')
plt.show()
```

```python
from mlxtend.plotting import plot_pca_correlation_graph
feature_names = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptration', 'b', 'lstat', 'medv']
figure, correlation_matrix = plot_pca_correlation_graph(data_scaled, 
                                                        feature_names,
                                                        dimensions=(1, 2),
                                                        figure_axis_size=10)
```

In a correlation circle:

- Each variable is represented as a vector (an arrow) in a multidimensional space.
- The direction of the vector indicates the nature of the correlation. If two vectors point in the same direction, the variables are positively correlated. If they point in opposite directions, the variables are negatively correlated.
- The length of the vector indicates the strength of the correlation. A longer vector implies a stronger correlation with other variables.
- The angle between the vectors represents the correlation between the variables. If the angle is small (acute), the correlation is positive. If the angle is large (obtuse), the correlation is negative. If the vectors are at a right angle, it means the variables are not correlated.

```python
principalDf1 = pd.DataFrame(data = principalComponents1[:, :2], columns = ['principal component 1', 'principal component 2'])

plt.figure(figsize=(8,6))
plt.scatter(principalDf1['principal component 1'], principalDf1['principal component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 component PCA')
plt.show()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)


pca = PCA()
principalComponents2 = pca.fit_transform(x_scaled)

plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Boston Housing Dataset Explained Variance')
plt.show()

```

This plot tells us how much information (variance) can be attributed to each of the principal components.
